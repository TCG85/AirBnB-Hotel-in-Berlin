```python
import pandas as pd
import numpy as np
import re 

pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
#had to skip the first rows in order to read the file
df = pd.read_csv("results.csv", skiprows=3, sep=',')
```

```python
#delete every 2nd row of dataframe because its empty
df = df.iloc[1::2]
```

```python
#reset the index, keep only fresh one
df = df.reset_index()
df = df.drop(["index"],axis=1)
df
```

```python
#adjust column names

df = df.drop(["Der Preis beträgt 468 €"],axis=1)
df = df.rename(columns={"The Ritz-Carlton": "hotel_name", " Berlin":"area", "Mitte":"price"})
df
```

```python
df = df.append(pd.Series(["The Ritz-Carlton", "Mitte", "468 €"], index=df.columns ), ignore_index=True)
df
```

```python
df = df.sort_values(by=['hotel_name'])
df = df.reset_index(drop=True)
df
```

```python
# Remove the string from price series
df['price'].replace(regex=True,inplace=True,to_replace=r'\D',value=r'')
```

```python
df.to_csv(r'expedia_data_clean.csv')
#export as csv
```
